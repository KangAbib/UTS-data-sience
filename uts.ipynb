```python
import numpy as np
import pandas as pd
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data2 = data1[['KOMODITAS','Harga Terkini (Rp)']]
data2.head()
```

```python
x = data2['KOMODITAS']
y = data2['Harga Terkini (Rp)']

plt.scatter(x, y)
plt.xlabel('Komoditas')
plt.ylabel('Harga Terkini (Rp)')
plt.title('Diagram Titik')


plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
# Mengelompokkan data berdasarkan komoditas dan menghitung jumlahnya
gb_komoditas = data2.groupby('KOMODITAS')['Harga Terkini (Rp)'].sum()
print(gb_komoditas)

# Membuat diagram lingkaran dari hasil pengelompokan
gb_komoditas.plot(kind='pie', title='Diagram Lingkaran Harga Terkini per Komoditas', figsize=(10, 6), autopct='%1.1f%%')
plt.ylabel('')  # Menghilangkan label sumbu y
plt.show()
```

```python
x = data2['KOMODITAS']
y = data2['Harga Terkini (Rp)']

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.bar(x,y)
plt.xlabel('Komoditas')
plt.ylabel('Harga Terkini (Rp)')
plt.title('Diagram Batang Harga Terkini per Komoditas')
plt.show()
```
